```python
import numpy as np
import matplotlib.pyplot as plt

# Define the unit step function
def u(n):
    return np.where(n >= 0, 1, 0)

# Define x[n]
n = np.arange(0, 20)
x = u(n) - u(n - 6) + u(n - 10) - u(n - 14)

# Define h[n]
h = 0.6 ** n * u(n)

# Perform convolution to get y[n]
y = np.convolve(x, h)

# Plot h[n]
plt.figure(figsize=(10, 6))
plt.stem(n, h, basefmt=" ", use_line_collection=True)
plt.title('h[n] = 0.6^n u[n]')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid(True)
plt.show()

# Plot x[n]
plt.figure(figsize=(10, 6))
plt.stem(n, x, basefmt=" ", use_line_collection=True)
plt.title('x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()

# Plot y[n]
n_y = np.arange(0, len(y))
plt.figure(figsize=(10, 6))
plt.stem(n_y, y, basefmt=" ", use_line_collection=True)
plt.title('y[n] = x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid(True)
plt.show()
```
