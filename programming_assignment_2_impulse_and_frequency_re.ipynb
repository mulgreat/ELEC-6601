**a) Plot the Impulse Response**

The impulse response, `h(n)`, of a causal LTI system can be obtained by performing the inverse Z-transform of its transfer function, `H(z)`.

**Python Code:**

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import dimpulse, dlti

# Define transfer function coefficients
num = [1, -0.6]  # Numerator coefficients
den = [1, 1.2, -0.36]  # Denominator coefficients

# Create a discrete-time linear time-invariant system
system = dlti(num, den)

# Get the impulse response
t, h_n = dimpulse(system, n=50)  # n specifies the duration of the response

# Convert the output to a more convenient form
h_n = h_n[0]

# Plot the impulse response with markers and line edges
plt.stem(t, h_n, markerfmt='o', linefmt='b-', label='Impulse Response')
plt.xlabel('Time (n)')
plt.ylabel('h(n)')
plt.title('Impulse Response of the LTI System')
plt.grid(True)
plt.legend()
plt.show()
```

**b) Plot the Amplitude and Phase Frequency Responses**

The frequency response of an LTI system is characterized by its amplitude and phase responses. We can obtain them from the transfer function using the frequency response function.

**Python Code:**

```
from scipy.signal import freqz

# Obtain frequency response
w, h = freqz(num, den)

# Calculate amplitude (magnitude) and phase responses
mag_h = np.abs(h)  # Magnitude
phase_h = np.angle(h)  # Phase in radians

# Convert phase to degrees (optional)
phase_h_deg = np.rad2deg(phase_h)

# Plot amplitude and phase responses
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(w, mag_h, label='Magnitude')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.title('Magnitude Frequency Response')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(w, phase_h_deg, label='Phase')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('Phase (degrees)')
plt.title('Phase Frequency Response')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**c) What type of filter does it represent?**

By analyzing the impulse response and frequency response, you can determine the filter type. In this case, the filter exhibits characteristics of a **high-pass filter**. Here's how you can identify it:

- **Impulse response:** Non-zero initial value and decays over time, indicating a delayed high-frequency component.
- **Magnitude response:** Generally decreases as frequency increases, allowing high frequencies to pass through more easily.

**d) Plot the Poles and Zeros of the System**

The poles and zeros of a transfer function are crucial in understanding the system's stability and frequency response.

**Python Code:**

```
# Find poles and zeros from the transfer function coefficients
poles = np.roots(den)
zeros = np.roots(num)

# Plot poles and zeros on the z-plane
plt.figure()
unit_circle = plt.Circle((0, 0), 1, color='black', fill=False)
plt.gca().add_patch(unit_circle)
plt.axhline(y=0, color='black', linestyle='--')
plt.axvline(x=0, color='black', linestyle='--')

# Plot poles with 'x' and zeros with 'o'
plt.plot(poles.real, poles.imag, marker='x', label='Poles')
plt.plot(zeros.real, zeros.imag, marker='o', label='Zeros')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Poles and Zeros of the LTI System')
plt.grid(True)
plt.legend()
plt.show()
```
