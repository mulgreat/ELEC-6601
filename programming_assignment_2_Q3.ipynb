```python

#a
import numpy as np
import matplotlib.pyplot as plt

# Frequency range for plotting
omega = np.linspace(-np.pi, np.pi, 1000)

# Compute the frequency response H(e^jw)
numerator = 1 - 0.4 * np.exp(-1j * omega) + np.exp(-2j * omega)
H = numerator  # since denominator is 1

# Plot the magnitude and phase of H(e^jw)
plt.figure(figsize=(14, 6))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.plot(omega, np.abs(H))
plt.title('Frequency Response of H(z)')
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Magnitude')
plt.grid(True)

# Phase plot
plt.subplot(2, 1, 2)
plt.plot(omega, np.angle(H))
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
#b
import scipy.signal as signal

# Input sequences
n = np.arange(0, 201)
x1 = 4 * np.cos(3 * np.pi * n / 10)
x2 = 2 * np.cos(6 * np.pi * n / 10)
x = x1 + x2

# System coefficients from H(z)
b = [1, -0.4, 1]
a = [1]

# Compute the output using the frequency response
y1 = signal.lfilter(b, a, x1)
y2 = signal.lfilter(b, a, x2)
y = signal.lfilter(b, a, x)

# Plot the results
plt.figure(figsize=(14, 8))

plt.subplot(3, 1, 1)
plt.plot(n, y1, label='Output y1(n)')
plt.title('Output for Input x1(n) = 4cos(3n/10)')
plt.xlabel('n')
plt.ylabel('y1(n)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(n, y2, label='Output y2(n)')
plt.title('Output for Input x2(n) = 2cos(6n/10)')
plt.xlabel('n')
plt.ylabel('y2(n)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(n, y, label='Output y(n)')
plt.title('Output for Input x(n) = x1(n) + x2(n)')
plt.xlabel('n')
plt.ylabel('y(n)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
#d
# Difference equation coefficients
b_coeffs = [1, -0.4, 1]
a_coeffs = [1]

# Compute the output using the difference equation
y_diff_eq = signal.lfilter(b_coeffs, a_coeffs, x)

# Plot the results
plt.figure(figsize=(14, 6))

plt.plot(n, y_diff_eq, label='Output using Difference Equation')
plt.title('Output for Input x(n) = x1(n) + x2(n) using Difference Equation')
plt.xlabel('n')
plt.ylabel('y(n)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
#e
# Check if the outputs match
output_difference = np.abs(y - y_diff_eq)

# Plot the difference
plt.figure(figsize=(14, 6))

plt.plot(n, output_difference, label='Difference between Methods')
plt.title('Difference between Outputs from Frequency Response and Difference Equation Methods')
plt.xlabel('n')
plt.ylabel('Difference')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
